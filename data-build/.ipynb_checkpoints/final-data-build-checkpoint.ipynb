```r
# Load functions

library(tidyverse)
```

```r
# Build final data set for 2014-2019

dat.2010 <- read.csv("../data/output/data-2010.csv")
dat.2011 <- read.csv("../data/output/data-2011.csv")
dat.2012 <- read.csv("../data/output/data-2012.csv")
dat.2013 <- read.csv("../data/output/data-2013.csv")
dat.2014 <- read.csv("../data/output/data-2014.csv")
dat.2015 <- read.csv("../data/output/data-2015.csv")
```

```r
# Merge

library(dplyr)

dfs <- list(
  `2010` = dat.2010,
  `2011` = dat.2011,
  `2012` = dat.2012,
  `2013` = dat.2013,
  `2014` = dat.2014,
  `2015` = dat.2015
)

ma.full <- bind_rows(dfs, .id = "year") %>%
  mutate(year = as.integer(year))

glimpse(ma.full)
```

```r
# Write CSV

write_csv(ma.full,"../data/output/ma-full.csv")
```

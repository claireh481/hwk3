## Question 5

```r
# Question 5

data.2010 <- read.csv("../data/output/data-2010.csv")

data_cleaned <- data.2010 %>%
  mutate(
    raw_score = rowMeans(
      as.matrix(select(., breastcancer_screen:ttyt_available)),
      na.rm = TRUE
    ),
    star_from_raw = floor(raw_score * 2 + 0.5) / 2,
    rounded_up = (star_from_raw > floor(raw_score * 2) / 2)
  )

rounded_up_table <- data_cleaned %>%
  filter(star_from_raw %in% c(3, 3.5, 4, 4.5, 5)) %>%
  group_by(star_from_raw) %>%
  summarize(
    num_plans_rounded_up = sum(rounded_up, na.rm = TRUE),
    .groups = "drop"
  ) %>%
  complete(star_from_raw = c(3, 3.5, 4, 4.5, 5),
           fill = list(num_plans_rounded_up = 0)) %>%
  arrange(star_from_raw) %>%
  rename(
    `Star Rating` = star_from_raw,
    `Plans Rounded Up` = num_plans_rounded_up
  )

rounded_up_table
```

## Question 6

```r
data_cleaned <- data_cleaned %>%
  mutate(
    market_share = avg_enrollment / avg_enrolled
  )

star30 <- lm(
  market_share ~ treat * score,
  data = data_cleaned %>%
    filter(raw_score >= (2.75 - 0.125),
           raw_score <= (2.75 + 0.125),
           Star_Rating %in% c(2.5, 3.0)) %>%
    mutate(
      treat = as.integer(raw_score >= 2.75),
      score = raw_score - 2.75
    )
)

star35 <- lm(
  market_share ~ treat * score,
  data = data_cleaned %>%
    filter(raw_score >= (3.25 - 0.125),
           raw_score <= (3.25 + 0.125),
           Star_Rating %in% c(3.0, 3.5)) %>%
    mutate(
      treat = as.integer(raw_score >= 3.25),
      score = raw_score - 3.25
    )
)


rd_table <- bind_rows(
  tidy(star30) %>%
    select(term, estimate) %>%
    mutate(Cutoff = "3 Stars"),

  tidy(star35) %>%
    select(term, estimate) %>%
    mutate(Cutoff = "3.5 Stars")
) %>%
  pivot_wider(
    names_from = Cutoff,
    values_from = estimate
  ) %>%
  mutate(
    term = recode(term,
      "(Intercept)" = "Intercept",
      "treat" = "Rounded",
      "score" = "Running Score",
      "treat:score" = "Rounded × Score"
    )
  ) %>%
  rename(Coefficient = term)

model_stats <- tibble(
  Coefficient = c("Observations", "R-squared"),
  `3 Stars` = c(nobs(star30), glance(star30)$r.squared),
  `3.5 Stars` = c(nobs(star35), glance(star35)$r.squared)
)

rd_table <- bind_rows(rd_table, model_stats)

rd_table <- rd_table %>%
  mutate(
    `3 Stars` = ifelse(Coefficient == "Observations",
                       round(`3 Stars`, 0),
                       round(`3 Stars`, 3)),
    `3.5 Stars` = ifelse(Coefficient == "Observations",
                         round(`3.5 Stars`, 0),
                         round(`3.5 Stars`, 3))
  )

rd_table
```

## Question 7

```r
library(tibble)

estimate_rd <- function(data, cutoff, bw) {

  rd <- data %>%
    filter(raw_score >= cutoff - bw,
           raw_score <= cutoff + bw) %>%
    filter(
      (cutoff == 2.75 & Star_Rating %in% c(2.5, 3.0)) |
      (cutoff == 3.25 & Star_Rating %in% c(3.0, 3.5))
    ) %>%
    mutate(
      treat = as.integer(raw_score >= cutoff),
      score = raw_score - cutoff
    )

  model <- lm(market_share ~ treat + score, data = rd)

  tibble(
    cutoff = cutoff,
    bandwidth = bw,
    estimate = unname(coef(model)["treat"]),
    se = summary(model)$coefficients["treat", "Std. Error"]
  )
}

bandwidths <- c(0.10, 0.12, 0.13, 0.14, 0.15)

cutoffs <- c(2.75, 3.25)

rd_results <- expand.grid(
  cutoff = cutoffs,
  bandwidth = bandwidths
) %>%
  rowwise() %>%
  do(estimate_rd(data_cleaned, .$cutoff, .$bandwidth)) %>%
  ungroup() %>%
  mutate(
    ci_low = estimate - 1.96 * se,
    ci_high = estimate + 1.96 * se,
    cutoff_label = factor(cutoff,
                          levels = c(2.75, 3.25),
                          labels = c("3-star", "3.5-star"))
  )

ggplot(rd_results, aes(x = bandwidth, y = estimate, color = cutoff_label)) +
  geom_point(size = 2) +
  geom_errorbar(aes(ymin = ci_low, ymax = ci_high), width = 0.003) +
  labs(
    title = "RD Estimates Across Different Bandwidths",
    x = "Bandwidth",
    y = "RD Estimate (Effect on Market Share)",
    color = "Cutoff"
  ) +
  theme_minimal()
```

My results are somewhat sensitive to the choice of bandwidth, tending to increase slightly (on average) as the bandwidth increases. The effect of receiving a 3-star rating on enrollments is around 0.05, but increases to nearly 0.10 at higher bandwidths. The effect of receiving a 3.5-star rating on market share is negative for all bandwidth values, but increases slightly as bandwidth increases.

## Question 9

```r
bw <- 0.125
cutoffs <- c(`3 stars` = 2.75, `3.5 stars` = 3.25)

data_cleaned <- data_cleaned %>%
  mutate(
    HMO = as.integer(grepl("HMO", plan_type)),
    PartD = as.integer(partd == "Yes")
  )

balance_simple <- purrr::imap_dfr(cutoffs, function(cut, label){

  d <- data_cleaned %>%
    filter(is.finite(raw_score),
           raw_score >= cut - bw,
           raw_score <= cut + bw) %>%
    mutate(side = if_else(raw_score >= cut, "Above", "Below"))
  means <- d %>%
    group_by(side) %>%
    summarize(
      HMO = mean(HMO, na.rm = TRUE),
      PartD = mean(PartD, na.rm = TRUE),
      n = n(),
      .groups = "drop"
    ) %>%
    select(-n) %>%
    pivot_longer(cols = c(HMO, PartD), names_to = "Characteristic", values_to = "Mean")

  diff <- means %>%
    pivot_wider(names_from = side, values_from = Mean) %>%
    mutate(
      Cutoff = label,
      Diff = Above - Below
    ) %>%
    select(Cutoff, Characteristic, Diff)

  diff
})

balance_simple

ggplot(balance_simple, aes(x = Diff, y = Characteristic, color = Cutoff)) +
  geom_vline(xintercept = 0, linetype = "dashed") +
  geom_point(size = 2) +
  labs(
    title = "Difference in Mean Characteristics Across Cutoffs",
    x = "Mean Differences",
    y = ""
  ) +
  theme_minimal()
```

The table and love plot indicate that around 40% more plans are HMOs at the 3-star threshold vs 2.5-stars. Also, plans just above the 3.0 threshold are more likely to have Part D. In contrast, the differences for the 3.5 star threshold are negative but smaller in magnitude. Plans just above the threshold value for 3.5 stars are slightly less likely to be HMOs or to have Part D. This suggests some differences in HMO and Part D status near the 3-star threshold and slight evidence of differences at the 3.5-star threshold.
